```python
#  Importar Pacotes
import pandas as pd
import numpy as np
import re

# Funções
def comeca_com_numero(texto):
  """Verifica se a string começa com um número."""
  return bool(re.match(r'^\d', texto))

# Caminho Até o Diretório
path = '/content/sample_data/'
filename =  'OUROCARD_ELO_NANQUIM-Fev_25.txt'
```

```python
# Ler Arquivo (data frame)
df = pd.read_csv(path + filename, delimiter='\t', encoding='latin-1', skiprows=47, skipfooter=59, engine='python')

# Nomear Coluna
df.columns = ['Coluna']
```

```python
# Acessar a coluna 'Coluna'
coluna = df['Coluna']

# Remover Linhas sem Data
df = df[df['Coluna'].apply(comeca_com_numero)]

# Dividir em Colunas
df['Data Mov.'] = coluna.str[:10]  # Primeiros 10 caracteres
df['Histórico'] = coluna.str[10:60] # Caracteres do 10º ao 60º (50 caracteres)
df['Valor'] = coluna.str[60:70] # Caracteres do 60º ao 81º (21 caracteres)
sub_df=df.drop(columns=['Coluna'])

# Inserir Coluna Mês
sub_df.insert(1, 'Mês/Ano', '')

# Formatar Data Mov.
data_formatada = sub_df['Data Mov.'].str.replace('.', '/', regex=False)
sub_df['Data Mov.'] = pd.to_datetime(data_formatada, format='%d/%m/%Y')
sub_df['Data Mov.'] = sub_df['Data Mov.'].dt.strftime('%d/%m/%Y')

# Formatar Mês/Ano
data_formatada = pd.to_datetime(data_formatada, format='%d/%m/%Y', errors='coerce')
sub_df['Mês/Ano'] = data_formatada.dt.strftime('%b/%y').str.capitalize()

# Adicionar coluna
sub_df['Tp'] = ''

# Substituir vírgula por ponto em Valor e converter para float
sub_df['Valor'] = sub_df['Valor'].str.replace('.', '', regex=False)
sub_df['Valor'] = sub_df['Valor'].str.strip().str.replace(',', '.', regex=False)

```

```python

# Atribuir Tp e Substituir ponto por vírgula em valor
sub_df['Tp'] = np.where(sub_df['Valor'].astype(float) < 0, 'C', 'D')
sub_df['Valor'] = sub_df['Valor'].str.strip().str.replace('.', ',', regex=False)

sub_df
```

```python
# Salvar e Exportar
sub_df.to_csv(path + filename.replace('.csv', '_formatado.csv'), sep=';', index=False)
```
